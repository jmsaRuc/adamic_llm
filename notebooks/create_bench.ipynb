```python
def sample_df(df, group_col, n_samples):
    frac = n_samples / len(df)
    df_ = df.groupby(group_col, group_keys=False).apply(lambda x: x.sample(frac=frac))
    df_[group_col] = df[df.index.isin(df_.index)][group_col]
    return df_
```

```python
for lang in mmmlu_name:
    ds = load_from_disk(f"../benchmarks/datasets/mmmlu_full/{lang}")
    df = ds.to_pandas()
    df_sampled = sample_df(df, "Subject", 500)
    ds_sampled = Dataset.from_pandas(df_sampled.reset_index(drop=True))
    ds_sampled.save_to_disk(f"../benchmarks/datasets/mmmlu/{lang}")
```
